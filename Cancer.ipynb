ID пациентов и пустой 32ой столбец не нужны, так как не содержат релевантной информации (скорее всего, вид опухоли не зависит от ФИО, даты диагноза и тд, а только от характеристик клеток опухоли). Их следует убрать.

```python
df.drop(columns = 'Unnamed: 32', axis = 1, inplace = True)

df.drop(columns = 'id', axis = 1, inplace = True)
```

```python
df.info()
```

Таким образом, на данный момент признаковое описание имеет размерность 29, целевое значение одно - разделение на классы M или B, таким образом, стоит задача двоичной классификации. 

Превратим названия классов в более удобные численные эквиваленты.

```python
from sklearn.preprocessing import LabelEncoder

L = LabelEncoder()

df['diagnosis'] = L.fit_transform(df['diagnosis'])
```
