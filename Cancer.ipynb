В будущем, интересным было бы посмотреть на поведение моделей на измененном признаковом описании - например, оставить только данные радиуса и площади или создать признак, являющийся линейной или степенной комбинацией данных признаков.

В дальнейшем, будем использовать StandardScaler для маштабирования данных, так как диапазоны некоторых признаков разительно отличаются друг от друга.

```python
X = df.drop('diagnosis',axis=1)

Y = df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_s = scaler.fit_transform(X)
```

```python
#kmeans

from sklearn.cluster import KMeans

t1 = X_s

kmeans = KMeans( n_clusters = 2, init='k-means++' ).fit_predict(X_s)
```

```python
Y
```

```python
y = Y.to_numpy()
```

```python
from sklearn.metrics import accuracy_score

print( accuracy_score(y, kmeans) )
```

```python
from sklearn.metrics import confusion_matrix

cm0 = confusion_matrix(y, kmeans)
```

```python
import seaborn as sns

sns.heatmap(cm0, annot = True, fmt = '.1f')
```

```python
import numpy as np
```

```python
acc = []

for i in range(50):
    
    kmeans = KMeans( n_clusters = 2, init = 'k-means++' ).fit_predict(X_s)
    
    temp = accuracy_score(y, kmeans)
    
    if temp < 0.50:
        
        acc.append(1 - temp)
        
    else:
        
        acc.append(temp)
    
acc = np.array(acc)

print(np.mean(acc))

print(np.min(acc))

print(np.max(acc))
```

```python
from sklearn.model_selection import train_test_split

xtr, xte, ytr, yte = train_test_split(X_s, Y, test_size = 0.75)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier()

clf_dt.fit(xtr, ytr)

ypre_dt = clf_dt.predict(xte)
```

```python
accuracy_score(yte, ypre_dt)
```

```python
cm1 = confusion_matrix(yte, ypre_dt)
```

```python
sns.heatmap(cm1, annot = True, fmt = '.1f')
```

```python
from sklearn.neural_network import MLPClassifier

clf_nn = MLPClassifier(solver='lbfgs', activation = 'relu', alpha = 1e-5, hidden_layer_sizes = (5, 2), random_state = 1)

#используем lbfgs, потому что он лучше работает на маленьких датасетах

#гиперпараметры были подобраны вручную

clf_nn.fit(xtr,ytr)

ypre_nn = clf_nn.predict(xte)
```

```python
accuracy_score(yte, ypre_nn)
```

```python
cm2 = confusion_matrix(yte, ypre_nn)
```

```python
sns.heatmap(cm2, annot = True, fmt='.1f')
```
